```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
```

```python
class Preprocessing:
    from sklearn.preprocessing import OrdinalEncoder
    from sklearn.pipeline import Pipeline
    from sklearn.preprocessing import OneHotEncoder
    from sklearn.preprocessing import MinMaxScaler
    from sklearn.compose import ColumnTransformer
    from sklearn.impute import SimpleImputer
        
    def __init__(self, data):
        self.data = data
    
    def preprocess_all(self, scale_var, encode_var, hot_var):
        num = self.num_pipeline()
        encode = self.encode_scale_pipeline()
        one_hot = self.one_hot_pipeline()
        
        pipeline = ColumnTransformer([
        ('one_hot', one_hot, hot_var),
        ('encode', encode, encode_var),
        ('num', num, scale_var)
        ], remainder='passthrough')
        
        data_final = pipeline.fit_transform(self.data)
        return data_final
    
    def encode_scale_pipeline(self):
        encode_pipeline = Pipeline([
            ('encode', OrdinalEncoder()),
            ('scale', MinMaxScaler())
        ])
        return encode_pipeline
    
    def encode_pipeline(self):
        encode_pipeline = Pipeline([
            ('encode', OrdinalEncoder())
        ])
        return encode_pipeline    
    
    def num_pipeline(self):
        num_pipeline = Pipeline([
            ('imputer', SimpleImputer(strategy="median")),
            ('minmax', MinMaxScaler())
        ])
        return num_pipeline
    
    def one_hot_pipeline(self):
        one_hot_pipeline = Pipeline([
            ('cat_one_hot', OneHotEncoder())
        ])
        return one_hot_pipeline   
```
